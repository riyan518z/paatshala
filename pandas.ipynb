```python
#5
Sc = Sc.rename(columns={'Title':'movie title'})
Sc
```

```python
#6
Sc=Sc['Genre'].unique()
Sc
```

```python
#7
Sc['Number_of_Genres'] = Sc['Genre'].str.count(",")+1
display(Sc[['Title', 'Genre', 'Number_of_Genres']])
```

```python
#8
Sc = Sc.dropna(subset=['Revenue (Millions)', 'Metascore'])
display(Sc[['Revenue (Millions)', 'Metascore']])
```

```python
#9
top5 = Sc.nlargest(5, 'Rating')[['movie title', 'Rating']]
print("Top 5 by Rating:\n", top5)
```

```python
#10
mvh = Sc.nlargest(1,'Revenue (Millions)')[['movie title', 'Revenue (Millions)']]
mvl =  Sc.nsmallest(1,'Revenue (Millions)')[['movie title', 'Revenue (Millions)']]
print(mvh)
print(mvl)
```

```python
#11
avg_rating_by_director = Sc.groupby('Director')['Rating'].mean().sort_values(ascending=False)
print("Average Rating by Director:\n")
avg_rating_by_director
```

```python
#16
Sc['Profitability']=Sc['Revenue (Millions)']/Sc['Runtime (Minutes)']
Sc
```

```python
#17
conditions=[Sc['Rating'] < 6 ,(Sc['Rating']>=6) & (Sc['Rating']<=7.5) ,Sc['Rating'] >=7.5 ]
values =["poor",'average','excellent']
Sc['movie_rating']= np.select(conditions,values,default='UNKNOWN')
Sc.head(10)
```

```python
#18
def calculate_high_score(row):

    return row['Rating'] * 10 + row['Revenue (Millions)']
Sc['High_Score'] = Sc.apply(calculate_high_score, axis=1)
print(Sc[['Title', 'Rating', 'Revenue (Millions)', 'High_Score']].head())
```
