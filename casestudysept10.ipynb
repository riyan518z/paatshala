```python
import pandas as pd
pop = pd.read_csv("/content/population.csv")
pop
```

```python
import pandas as pd
gdp = pd.read_excel("/content/gdp.xlsx")
gdp
```

```python
import pandas as pd
ius = pd.read_json("/content/internet_users.json")
ius
```

```python
import pandas as pd
lit = pd.read_xml("/content/literacy_rate.xml")
lit
```

```python
pop.rename(columns={'country':'name'}, inplace=True)
pop
```

```python
gdp.rename(columns={'Country':'name'}, inplace=True)
gdp
```

```python
ius.rename(columns={'country':'name'}, inplace=True)
ius
```

```python
lit
```

```python
pop_gdp = pd.merge(pop,gdp, on='name', how='inner')
pop_gdp
```

```python
pop_gdp_ius = pd.merge(pop_gdp,ius, on='name', how='inner')
pop_gdp_ius
```

```python
pop_gdp_ius_lit = pd.merge(pop_gdp_ius,lit, on='name', how='inner')
pop_gdp_ius_lit
```

```python
import pandas as pd


pop_gdp_ius_lit["internet_penetration_rate"] = (pop_gdp_ius_lit["internet_users"] / pop_gdp_ius_lit["population"] * 100)

pop_gdp_ius_lit
```

```python
high = pop_gdp_ius_lit.loc[pop_gdp_ius_lit['internet_penetration_rate'].idxmax()]
high
```

```python
avg = pop_gdp_ius_lit['literacy_rate'].mean()
print(avg)
```

```python
pop_gdp_ius_lit[['population','literacy_rate','internet_users']].corr()
```

```python
import sqlite3

conn = sqlite3.connect("merge")
cursor = conn.cursor()
pop_gdp_ius_lit.to_sql(name='merge', con=conn, if_exists='replace', index=False)
```

```python
cursor.execute("select * from merge")
rows = cursor.fetchall()
for row in rows:
  print(row)
```
