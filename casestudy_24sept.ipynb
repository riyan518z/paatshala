```python
num_house = new_house.select_dtypes(include='float64')
num_house
```

```python
import matplotlib.pyplot as plt

for i in num_house:
    plt.figure(figsize=(10,5))
    plt.boxplot(new_house[i])
    plt.title(i)
```

```python
import numpy as np

q1 = np.quantile(new_house['No of Bathrooms'],0.25)
q3 = np.quantile(new_house['No of Bathrooms'],0.75)

iqr = q3 - q1

upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)

print(upper_bound)
print(lower_bound)
print(iqr)
```

```python
bathroom_out = []
for i in new_house['No of Bathrooms'] :
  if i > upper_bound or i < lower_bound :
    bathroom_out.append(i)
```

```python
print(bathroom_out)
```

```python
new_house['No of Bathrooms'] = new_house['No of Bathrooms'].clip(lower=lower_bound,upper=upper_bound)
```

```python
plt.boxplot(new_house['No of Bathrooms'])
```
