```python
#3
import random
ra = np.random.randint(10, 101, size=15)
u = np.where(ra > 50, -1, ra)
c = np.sum(u == -1)

print(ra)
print(u)
print(c)
```

```python
#4
arr = np.random.rand(5, 4)
max_row_index = np.argmax(np.sum(arr, axis=1))

row_min = arr.min(axis=1, keepdims=True)
row_max = arr.max(axis=1, keepdims=True)
normalized = (arr - row_min) / (row_max - row_min)

print("\n4. Random 5x4 Array:\n", arr)
print("Index of Row with Max Sum:", max_row_index)
print("Normalized Array:\n", normalized)
```

```python
#5
arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
unique = np.unique(arr)

counts = np.bincount(arr)
duplicates = np.where(counts > 1)[0]

print("\n Original Array:\n", arr)
print("Unique Elements:\n", unique)
print("Duplicates:\n", duplicates)
```

```python
#6
arr = np.arange(1, 37).reshape(6, 6)
sub = arr[1:5, 1:5]
e = sub[sub % 2 == 0]

print("\n6. Original 6x6 Array:\n", arr)
print("Middle 4x4 Submatrix:\n", sub)
print("Even Numbers in Submatrix:\n", e)
```

```python
#7
arr = np.arange(12).reshape(3, 4)
re = arr.reshape(2, 6)
f = re.flatten()
reshaped_again = f.reshape(4, 3)

print("\n7. Original 3x4 Array:\n", arr)
print("Reshaped to 2x6:\n", re)
print("Flattened to 1D:\n", f)
print("Reshaped to 4x3:\n", reshaped_again)
```

```python
#8
a = np.array([[1, 2], [3, 4]])
newr = np.vstack([a, [[5, 6]]])
newc = np.hstack([newr, [[7], [8], [9]]])

print(f'array a:\n{a}\n')
print(f'array after appending row:\n{newr}\n')
print(f'array after appending row and column:\n{newc}')
```

```python
#9
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])
d = np.concatenate([a, b, c])
e = d.reshape((3, 3))
f = np.diag(e)

print(f'array a:\n{a}\n')
print(f'array b:\n{b}\n')
print(f'array c:\n{c}\n')
print(f'joined single array d:\n{d}\n')
print(f'reshaped 3X3 array e:\n{e}\n')
print(f'extracted diagonal elements are:\n{f}')
```

```python
#10
arr = np.arange(1, 21)
filtered = arr[(arr % 3 == 0) | (arr % 5 == 0)]
count = filtered.size

print("\n Original Array:\n", arr)
print("Values Divisible by 3 or 5:\n", filtered)
print("Count:", count)
```

```python
titanic['age_group'] = np.where(titanic['age'] < 18, 'Child', np.where(titanic['age'] <= 60, 'Adult', 'Senior'))
def classify_age(age):
  if age < 18:
    return 'Child'
  elif age <= 60:
    return 'Adult'
  else:
    return 'Senior'

  titanic['age_group'] = titanic['age'].apply(classify_age)
```
